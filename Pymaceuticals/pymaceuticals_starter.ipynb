### Analysis

- This study is a study of the effects of nine different drugs on mice.  One group of mice was given placebos to serve as a control group for the study.  49% of the mice tested were female and 51% were male.  For the mice treated with Capomulin, Ramicane, Infubinol, and Ceftamin, there was only one mouse that was an outlier in terms of its final tumor volume.  The results indicate that the final tumor volume for the mice were lower for mice treated with the drug regimen of Capomulin and Ramicane were lower than the other drugs.  The study showed a strong positive correlation between mice weight and it's average tumor size.
 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
#data_complete = pd.merge(mouse_metadata, study_results, how="outer", on=["Mouse ID", "Mouse ID"])
#data_complete = study_results.merge(mouse_metadata, how='right', on='Mouse ID')
data_complete = pd.merge(study_results, mouse_metadata, how="outer", on="Mouse ID")

# Display the data table for preview
data_complete
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
reduced_columns = removed.loc[:, ["Drug Regimen","Tumor Volume (mm3)"]]
renamed_df = reduced_columns.rename(columns = {"Tumor Volume (mm3)":"Tumor Volume"})

grouped_mean = renamed_df.groupby(by=["Drug Regimen"]).mean()
renamed_mean = grouped_mean.rename(columns = {"Tumor Volume":"Mean Tumor Volume"})
grouped_median = renamed_df.groupby(by=["Drug Regimen"]).median()
renamed_median = grouped_median.rename(columns = {"Tumor Volume":"Median Tumor Volume"})
grouped_variance = renamed_df.groupby(by=["Drug Regimen"]).var()
renamed_variance = grouped_variance.rename(columns = {"Tumor Volume":"Tumor Volume Variance"})
grouped_sd = renamed_df.groupby(by=["Drug Regimen"]).std()
renamed_sd = grouped_sd.rename(columns = {"Tumor Volume":"Tumor Volume Std. Dev."})
grouped_sem = renamed_df.groupby(by=["Drug Regimen"]).sem()
renamed_sem = grouped_sem.rename(columns = {"Tumor Volume":"Tumor Volume Std. Err."})


# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary DataFrame.

df5 = renamed_mean.merge(renamed_median,on='Drug Regimen').merge(renamed_variance,on='Drug Regimen').merge(renamed_sd,on='Drug Regimen').merge(renamed_sem,on='Drug Regimen')
df5
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin
cols = ['Mouse ID','Tumor Volume (mm3)','Drug Regimen','Timepoint']
cols1 = ['Mouse ID','Timepoint']
reduced_cols = removed.loc[:, cols]
# Start by getting the last (greatest) timepoint for each mouse
grouped_max = reduced_cols.groupby(by=["Mouse ID"]).max()
grouped_max = grouped_max.reset_index()
reduced_col1 = grouped_max.loc[:, cols1]
# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
df2 = reduced_col1.merge(reduced_cols,on=['Mouse ID','Timepoint'], how="left")
df2
```

```python
# Put treatments into a list for for loop (and later for plot labels)
#treatments per instructions
treatments = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']

# Create empty list to fill with tumor vol data (for plotting)
tumor_vol_data = []
for x in treatments:
    
    # Locate the rows which contain mice on each drug and get the tumor volumes
    df3 = df2.loc[df2['Drug Regimen'] == x, 'Tumor Volume (mm3)']
    
    # Calculate the IQR and quantitatively determine if there are any potential outliers. 
    quartiles = df3.quantile([.25,.5,.75])
    lowerq = quartiles[0.25]
    upperq = quartiles[0.75]
    iqr = upperq-lowerq  
    
    print(f"The lower quartile of tumor volume is: {lowerq} for {x}")
    print(f"The upper quartile of tumor volume is: {upperq} for {x}")
    print(f"The interquartile range of tumor volume is: {iqr} for {x}")
    print(f"The the median of tumor volume is: {quartiles[0.5]} for {x}")
        
    # add subset 
    tumor_vol_data.append(df3)
    
    # Determine outliers using upper and lower bounds
    lower_bound = lowerq - (1.5*iqr)
    upper_bound = upperq + (1.5*iqr)
    print(f"Values below {lower_bound} could be outliers for {x}.")
    print(f"Values above {upper_bound} could be outliers for {x}.")  
    
    # Find outliers for each treatment
    outlier = df3.loc[(df3 > upper_bound) | (df3 < lower_bound)]
    print(f"Outlier for {x} is {outlier}")
    print()
    print()
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.
outlier_marker = {"markerfacecolor": "red", "markersize": 12}
fig1, ax1 = plt.subplots()
ax1.set_ylabel('Final Tumor Volume (mm3)')
ax1.boxplot(tumor_vol_data, labels=treatments, flierprops=outlier_marker)
plt.show()
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a mouse treated with Capomulin
Capomulin1 = removed.loc[removed['Drug Regimen'] == 'Capomulin']
Capomulin2 = Capomulin1.loc[Capomulin1['Mouse ID'] == 'l509']
Capomulin3 = Capomulin2[["Timepoint","Tumor Volume (mm3)"]]

Capomulin3.plot(x="Timepoint", y="Tumor Volume (mm3)", legend=False)
plt.ylabel("Tumor Volume (mm3)")

plt.show()
```

```python
# Generate a scatter plot of average tumor volume vs. mouse weight for the Capomulin regimen
capomulin_average = Capomulin1.groupby(['Mouse ID']).mean()
plt.scatter(capomulin_average['Weight (g)'],capomulin_average['Tumor Volume (mm3)'])
plt.xlabel('Mouse Weight')
plt.ylabel('Average Tumor Volume (mm3)')
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen
weight = capomulin_average['Weight (g)']
ave_tumor_volume = capomulin_average['Tumor Volume (mm3)']
correlation = st.pearsonr(weight,ave_tumor_volume)
print(f"The correlation between mouse weight and the average tumor volume is {round(correlation[0],2)}")

from scipy.stats import linregress
x_values = weight
y_values = ave_tumor_volume
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.xlabel('Mouse Weight')
plt.ylabel('Average Tumor Volume (mm3)')
plt.show()
```
